```python
# Базовая сортировка пузырьком
bs = pd.read_csv('BubbleSort.csv', delimiter=';')
```

```python
# Вычет одного из второго цикла
bsJarray = pd.read_csv('BubbleSortJarray.csv', delimiter=';')
```

```python
# Оптимизация с уменьшением размера массива
bsOpt = pd.read_csv('BubbleSortOptimized.csv', delimiter=';')
```

```python
# Булевая проверка swap
bsBool = pd.read_csv('BubbleSortBool.csv', delimiter=';')
```

```python
cs = pd.read_csv('CocktailSort.csv', delimiter=';')
```

```python
qs = pd.read_csv('QuickSort.csv', delimiter=';')
```

```python
fig = plt.figure(dpi=600)
ax1 = fig.add_subplot(111)

ax1.plot(bs['x'], bs['y'], label='BS',color='r', linestyle='--', linewidth=0.7)
ax1.plot(bsJarray['x'], bsJarray['y'], label='BSJarray',color='b', linestyle='-', linewidth=0.7)
ax1.plot(bsOpt['x'], bsOpt['y'], label='BSOptimized',color='k', linestyle='-', linewidth=0.7)
ax1.plot(bsBool['x'], bsBool['y'], label='BSBool',color='g', linestyle='-', linewidth=0.7)
ax1.plot(cs['x'], cs['y'], label='Cocktail',color='m', linestyle='-', linewidth=0.7)
ax1.plot(qs['x'], qs['y'], label='QuickSort',color='y', linestyle='-', linewidth=0.8)

ax1.set_title("Тест модификаций BubbleSort")    
ax1.set_xlabel('Размер массива, шаг = 100')
ax1.set_ylabel('Время, ms')
ax1.grid()

leg = ax1.legend()
plt.savefig('BubbleSortTest.png', dpi = 600)
```
